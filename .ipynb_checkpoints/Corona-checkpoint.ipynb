```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
```

```python
url = "https://covidtracking.com/api/states/daily?state=TX"
usData = requests.get(url=url).json()
df = pd.DataFrame(usData)
df.shape
```

```python
df.columns
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['days_since_outbreak'] = pd.Series(np.arange(df.shape[0] - 1, -1, -1))
df.head()
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.bar(df['date'], df['negative'], color='green')
ax.bar(df['date'], df['positive'], color='red')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.bar(df['date'], df['negative'], color='green')
ax.bar(df['date'], df['positive'], color='red')
plt.yscale('log')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.bar(df['date'], df['positive'], color='red')
ax.plot(df['date'], df['positive'], color='green')

ax.set(xlabel="Date",
       ylabel="Positive Cases",
       title="Total Positive Cases of COVID-19 in Texas")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(df['date'], df['positiveIncrease'], color='red')
ax.plot(df['date'], df['positive'], color='green')
ax.set(xlabel="Date",
       ylabel="Case Increase",
       title="Increase of Cases by day of COVID-19 in Texas")
plt.yscale('log')
plt.show()
```

```python
import seaborn as sns

sns.regplot(x='days_since_outbreak', y='positiveIncrease', data=df, order=3)
```

```python
from lmfit.models import PolynomialModel

model = PolynomialModel(degree=5)
params = model.guess(df['positive'], df['days_since_outbreak'])

result = model.fit(df['positive'], params, x=df['days_since_outbreak'])

result.plot_fit()
plt.show()

print(result.fit_report())
```

```python
f = np.polyfit(df['days_since_outbreak'], df['positive'], 5)
p = np.poly1d(f)
print(p)
```

```python
x = np.array(range(0, 41))  
formula = '-0.002392*x**5 + 0.1712*x**4 - 3.8*x**3 + 33.17*x**2 - 97.61*x + 50.68'
y = eval(formula)
plt.plot(x, y) 
plt.plot(df['days_since_outbreak'], df['positive'])
plt.title('Extrapolated Cases of COVID-19')
plt.show()
```

```python
url = "https://covidtracking.com/api/states/daily?state=TX"
usData = requests.get(url=url).json()
texas = pd.DataFrame(usData)
texas.shape
```

```python
url = "https://covidtracking.com/api/states/daily?state=NY"
usData = requests.get(url=url).json()
york = pd.DataFrame(usData)
york.shape
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(texas['date'], texas['positive'], color='red')
ax.plot(york['date'], york['positive'], color='green')
ax.set(xlabel="Date",
       ylabel="Case Increase",
       title="Cases in New York versus Texas")
plt.yscale('log')
plt.show()
```
